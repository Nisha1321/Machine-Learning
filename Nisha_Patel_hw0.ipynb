## Part 4: Getting started with File Input & Output (20 points)

Python can easily handle reading and writing files. As far as Python is concerned, a file is just a string (often very large!) stored on your local file system, that you can read or write, gradually or all together.

### Writing Files

*open()* returns a file object, and is most commonly used with two arguments: *open(filename, mode)*.

```python
outFile = open('sample.txt', 'w')
outFile.write('The first line\nThe second line\nThe third line\n')
outFile.close()
```

The first line creates a file object, which links Python to your computers file system. The first parameter in the file constructor gives the file name, *sample.txt*. The second parameter indicates how you use the file. The 'w' is short for **w**rite. If the file already existed, the old contents are destroyed.

And 'a' opens the file for **a**ppending; any data written to the file is automatically added to the end. 'r+' opens the file for both reading and writing.

The *close* line is essential for Python to make sure everything is really written, and to relinquish control of the file.

### Methods of File Objects

To read a files contents, call *f.read(size)*, which reads some quantity of data and returns it as a string. *size* is an optional numeric argument. When *size* is omitted or negative, the entire contents of the file will be read and returned.

```python
outFile = open('sample.txt', 'r')
outFile.read()
```

For reading lines from a file, you can loop over the file object. This is memory efficient, fast, and leads to simple code:

```python
outFile = open('sample.txt', 'r')
for line in outFile:
    print(line, end='')
```

It is good practice to use the *with* keyword when dealing with file objects. This has the advantage that the file is properly closed after its suite finishes, even if an exception is raised on the way. It is also much shorter than writing equivalent try-finally blocks:

```python
with open('sample.txt', 'r') as f:
    outFile = f.read()
outFile
```

## Part 5: Counting Word Frequency (20 points)

In the following example, we will try a fun task to count the word frequency in a document.

we start with a string and split it into a list. We then create an (initially empty) list called *wordfreq*, go through each word in the *wordlist*, and count the number of times that word appears in the whole list. We then add each words count to our wordfreq list. Using the *zip* operation, we are able to match the first word of the word list with the first number in the frequency list, the second word and second frequency, and so on. We end up with a list of word and frequency pairs. The *str* function converts any object to a string so that it can be printed.

```python
wordstring = 'it was the best of times it was the worst of times '
wordstring += 'it was the age of wisdom it was the age of foolishness'

wordlist = wordstring.split()

wordfreq = []
for w in wordlist:
    wordfreq.append(wordlist.count(w))

print("String\n" + wordstring +"\n")
print("List\n" + str(wordlist) + "\n")
print("Frequencies\n" + str(wordfreq) + "\n")
print("Pairs\n" + str(dict(zip(wordlist, wordfreq))))
```

Now, your job is to open up the url https://zziwei.github.io/CS584/hw0.txt we mentioned a few cells above. Then output the word-frequency pairs in that url, sorted in order of descending frequency. In the case of ours Dickens example here, the output would look something like: [('of', 4), ('it', 4), ('the', 4), ('was', 4), ('age', 2), ('times', 2), ('foolishness', 1), ('worst', 1), ('wisdom', 1), ('best', 1)].

Your job is to output a the word-frequency pairs for the url.
**Hint: use the built-in function 'translate' to remove all punctuation in the text**

```python
# your code here

import requests  # the lib that handles the url request
import string

url = 'https://zziwei.github.io/CS584/hw0.txt'
response = requests.get(url)
data = response.text

translator = str.maketrans('','',string.punctuation)
clean_data = data.translate(translator)

print(clean_data)
print()

datalist = clean_data.split()
WordFreq = []
for w in datalist:
    WordFreq.append(datalist.count(w))



print("Descending pair according to word count\n", 
      sorted(dict(zip(datalist, WordFreq)).items(), key=lambda x: x[1], reverse=True))

```
